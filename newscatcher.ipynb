```python
# print('cnn.com' in sources_and_latest['source'])
# print((sources_and_latest['source'].eq('cnn.com')).any())
# sources_and_latest['source']
import re
new_df = []
res = []
for index, row in df_with_sources.iterrows():
    # separating for multiple names
    source = row['news_source'].split(' ')
    name = row['url_name'].split('-') # separating for multiple names

    # check firt name .com and first name.org and source .org or source .com
    name_com = name[0] + '.com'
    name_org = name[0] + '.org'
    source_com = source[0].lower() + '.com'
    source_org = source[0].lower() + '.org'

    if (sources_and_latest['source'].eq(name_com)).any():
        res.append(name_com)
        new_df.append(row.tolist())

    if (sources_and_latest['source'].eq(name_org)).any():
        res.append(name_org)
        new_df.append(row.tolist())

    if (sources_and_latest['source'].eq(source_com)).any():
        res.append(source_com)
        new_df.append(row.tolist())

    if (sources_and_latest['source'].eq(source_org)).any():
        res.append(source_org)
        new_df.append(row.tolist())
    
    # check if source stripped has more than one word
    if len(name) > 1:
        
        # check if full title .com or .org
        names = ''.join(name)
        names_com = names + '.com'
        names_org = names + '.org'
        if (sources_and_latest['source'].eq(names_com)).any():
            res.append(names_com)
            new_df.append(row.tolist())
        
        if (sources_and_latest['source'].eq(names_org)).any():
            res.append(names_org)
            new_df.append(row.tolist())

        # check for initials .com and initials.org
        initials = [x[0] for x in name]
        initials = ''.join(initials)
        initials_com = initials + '.com'
        initials_org = initials +'.org'

        if (sources_and_latest['source'].eq(initials_com)).any():
            res.append(initials_com)
            new_df.append(row.tolist())

        if (sources_and_latest['source'].eq(initials_org)).any():
            res.append(initials_org)
            new_df.append(row.tolist())
        
        # check for full name with 'the' in the begging (like 'thepoliticalinsider.com' instead of 'politicalinsider.com')
        the_names_com = 'the' + names + '.com'
        the_names_org = 'the' + names + '.org'

        if (sources_and_latest['source'].eq(the_names_com)).any():
            res.append(the_names_com)
            new_df.append(row.tolist())
        
        if (sources_and_latest['source'].eq(the_names_org)).any():
            res.append(the_names_org)
            new_df.append(row.tolist())
        
        # check for initials with 'the' in the begging (like 'thepoliticalinsider.com' instead of 'politicalinsider.com')
        # which in turn would be 'tpi.com' instead of 'pi.com'
        the_initials_com = 't' + initials_com
        the_initials_org = 't' + initials_org

        if (sources_and_latest['source'].eq(the_initials_com)).any():
            res.append(the_initials_com)
            new_df.append(row.tolist())
        
        if (sources_and_latest['source'].eq(the_initials_org)).any():
            res.append(the_initials_org)
            new_df.append(row.tolist())
        

        # same checking for sources

        # check if full title .com or .org
        sources = ''.join(source).lower()
        sources_com = names + '.com'
        sources_org = names + '.org'
        if (sources_and_latest['source'].eq(sources_com)).any():
            res.append(sources_com)
            new_df.append(row.tolist())
        
        if (sources_and_latest['source'].eq(sources_org)).any():
            res.append(sources_org)
            new_df.append(row.tolist())
        
        # check for source initials .com and initials.org
        sources_initials = [x[0] for x in source]
        sources_initials = ''.join(sources_initials)
        sources_initials_com = sources_initials + '.com'
        sources_initials_org = sources_initials +'.org'

        if (sources_and_latest['source'].eq(sources_initials_com)).any():
            res.append(sources_initials_com)
            new_df.append(row.tolist())

        if (sources_and_latest['source'].eq(sources_initials_org)).any():
            res.append(sources_initials_org)
            new_df.append(row.tolist())

        # check for full name with 'the' in the begging (like 'thepoliticalinsider.com' instead of 'politicalinsider.com')
        the_sources_com = 'the' + sources + '.com'
        the_sources_org = 'the' + sources + '.org'

        if (sources_and_latest['source'].eq(the_sources_com)).any():
            res.append(the_sources_com)
            new_df.append(row.tolist())
        
        if (sources_and_latest['source'].eq(the_sources_org)).any():
            res.append(the_sources_org)
            new_df.append(row.tolist())
        
        # check for source initials with 'the' in the begging (like 'thepoliticalinsider.com' instead of 'politicalinsider.com')
        # which in turn would be 'tpi.com' instead of 'pi.com'
        the_sources_initials_com = 't' + sources_initials_com
        the_sources_initials_org = 't' + sources_initials_org

        if (sources_and_latest['source'].eq(the_sources_initials_com)).any():
            res.append(the_sources_initials_com)
            new_df.append(row.tolist())
        
        if (sources_and_latest['source'].eq(the_sources_initials_org)).any():
            res.append(the_sources_initials_org)
            new_df.append(row.tolist())
        
        # check if title ends in 'news'
        if name[-1] == 'news':

            # check if full title without news .com or .org
            names_wo_news = name[:-1]
            names_wo_news = ''.join(names_wo_news)

            names_wo_news_com = names_wo_news + '.com'
            names_wo_news_org = names_wo_news + '.org'

            if (sources_and_latest['source'].eq(names_wo_news_com)).any():
                res.append(names_wo_news_com)
                new_df.append(row.tolist())
            
            if (sources_and_latest['source'].eq(names_wo_news_org)).any():
                res.append(names_wo_news_org)
                new_df.append(row.tolist())

            # check if initials without news .com or .org
            initials_wo_news = [x[0] for x in names_wo_news]
            initials_wo_news = ''.join(initials_wo_news)

            initials_wo_news_com = initials_wo_news + '.com'
            initials_wo_news_org = initials_wo_news +'.org'

            if (sources_and_latest['source'].eq(initials_wo_news_com)).any():
                res.append(initials_wo_news_com)
                new_df.append(row.tolist())

            if (sources_and_latest['source'].eq(initials_wo_news_org)).any():
                res.append(initials_wo_news_org)
                new_df.append(row.tolist())
            
            # check for full name with 'the' in the beggining without news
            the_names_wo_news_com = 'the' + names_wo_news + '.com'
            the_names_wo_news_org = 'the' + names_wo_news + '.org'

            if (sources_and_latest['source'].eq(the_names_wo_news_com)).any():
                res.append(the_names_wo_news_com)
                new_df.append(row.tolist())
            
            if (sources_and_latest['source'].eq(the_names_wo_news_org)).any():
                res.append(the_names_wo_news_org)
                new_df.append(row.tolist())
            
            # check for initials with 'the' in the begging (like 'thepoliticalinsider.com' instead of 'politicalinsider.com')
            # which in turn would be 'tpi.com' instead of 'pi.com'
            the_initials_wo_news_com = 't' + initials_wo_news_com
            the_initials_wo_news_org = 't' + initials_wo_news_org

            if (sources_and_latest['source'].eq(the_initials_wo_news_com)).any():
                res.append(the_initials_wo_news_com)
                new_df.append(row.tolist())
            
            if (sources_and_latest['source'].eq(the_initials_wo_news_org)).any():
                res.append(the_initials_wo_news_org)
                new_df.append(row.tolist())
        
        # check if source ends in 'news'
        if source[-1] == 'news':

            # check if full title without news .com or .org
            sources_wo_news = source[:-1]
            sources_wo_news = ''.join(sources_wo_news)

            sources_wo_news_com = sources_wo_news + '.com'
            sources_wo_news_org = sources_wo_news + '.org'

            if (sources_and_latest['source'].eq(sources_wo_news_com)).any():
                res.append(sources_wo_news_com)
                new_df.append(row.tolist())
            
            if (sources_and_latest['source'].eq(sources_wo_news_org)).any():
                res.append(sources_wo_news_org)
                new_df.append(row.tolist())

            # check if initials without news .com or .org
            sources_initials_wo_news = [x[0] for x in sources_initials_wo_news]
            sources_initials_wo_news = ''.join(sources_initials_wo_news)

            sources_initials_wo_news_com = sources_initials_wo_news + '.com'
            sources_initials_wo_news_org = sources_initials_wo_news +'.org'

            if (sources_and_latest['source'].eq(sources_initials_wo_news_com)).any():
                res.append(sources_initials_wo_news_com)
                new_df.append(row.tolist())

            if (sources_and_latest['source'].eq(sources_initials_wo_news_org)).any():
                res.append(sources_initials_wo_news_org)
                new_df.append(row.tolist())
            
            # check for full name with 'the' in the beggining without news
            the_sources_wo_news_com = 'the' + sources_wo_news + '.com'
            the_sources_wo_news_org = 'the' + sources_wo_news + '.org'

            if (sources_and_latest['source'].eq(the_sources_wo_news_com)).any():
                res.append(the_sources_wo_news_com)
                new_df.append(row.tolist())
            
            if (sources_and_latest['source'].eq(the_sources_wo_news_org)).any():
                res.append(the_sources_wo_news_org)
                new_df.append(row.tolist())
            
            # check for initials with 'the' in the begging (like 'thepoliticalinsider.com' instead of 'politicalinsider.com')
            # which in turn would be 'tpi.com' instead of 'pi.com'
            the_sources_initials_wo_news_com = 't' + sources_initials_wo_news_com
            the_sources_initials_wo_news_org = 't' + sources_initials_wo_news_org

            if (sources_and_latest['source'].eq(the_sources_initials_wo_news_com)).any():
                res.append(the_sources_initials_wo_news_com)
                new_df.append(row.tolist())
            
            if (sources_and_latest['source'].eq(the_sources_initials_wo_news_org)).any():
                res.append(the_sources_initials_wo_news_org)
                new_df.append(row.tolist())
print(res)
print(len(res))
print(new_df)
print(len(new_df))
```
